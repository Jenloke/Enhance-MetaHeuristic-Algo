```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import numpy as np
import pandas as pd
import ast
```

```python
table_algo = pd.read_csv('output.csv')
table_algo.info()
```

```python
ast_native = ['solArray', 'best_SolutionPerIteration', 'decrepancy']

for change in ast_native:
	table_algo.loc[:, change] = table_algo[change].map(lambda x: ast.literal_eval(x))
```

```python
sets = [1, 2, 3]
sizes = [100, 200, 500, 1000, 2000, 5000]
iterations = [100, 500]

table_convergenceSpeed = table_algo[['set', 'size', 'iterations', 'algo', 'best_SolutionPerIteration']]
```

```python
# Objective 1
algo = ['test_hsa', 'test_pso']

for itr in iterations:
	for set in sets:
		for size in sizes:
			graphtable = table_convergenceSpeed.loc[(table_convergenceSpeed['set'] == set) & (table_convergenceSpeed['size'] == size) & (table_convergenceSpeed['iterations'] == itr)]
			# fig = plt.figure()
			fig, ax = plt.subplots(figsize=(5,3), layout='constrained')
			ax.xaxis.set_major_locator(MaxNLocator(integer=True))
			ax.yaxis.set_major_locator(MaxNLocator(integer=True))
			
			for a in algo:
				graph_algo = graphtable.loc[graphtable['algo'] == a]
				graph = graph_algo.iloc[0]['best_SolutionPerIteration']
				length = len(graph)
				
				# ax.set_ylim(0, max(graph_algo.iloc[0]['best_SolutionPerIteration']))
    
				if a == 'test_hsa':
					plt.plot(range(length), graph, linestyle='--', color='cyan', label='HSA', alpha=0.6)
				if a == 'test_pso':
					plt.plot(range(length), graph, linestyle='-', color='green', label='PSO', alpha=0.6)

			problem = f'Set {set}, Size {size}, Itr {itr}'

			plt.ylabel('Knapsack Value')
			plt.xlabel('Iteration')
			ax.set_title(problem, loc='left')
			ax.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc='lower right', ncols=1, borderaxespad=0.)
			plt.show()
```
